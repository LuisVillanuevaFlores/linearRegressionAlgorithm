```python
import numpy as  np
import matplotlib.pyplot as plt
```

```python
def estimate_b0_b1(x,y):
  n=np.size(x)
  #se calcula las medias de x y y
  m_x,m_y= np.mean(x), np.mean(y)

  #se calcula la sumatoria de XY y la de XX.
  Sumatoria_xy=np.sum((x-m_x)*(y-m_y))
  Sumatoria_xx=np.sum((x-m_x)**2)

  #se calcula los coeficientes de regresion osea b0 y b1

  b_1=Sumatoria_xy / Sumatoria_xx
  b_0= m_y - b_1*m_x
  return (b_0,b_1)
```

```python
#Funcion para graficado
def plot_regression(x,y,b):
  plt.scatter(x,y,color="g",marker="o",s=30)
  y_pred=b[0]+b[1]*x
  plt.plot(x,y_pred,color="b")

  plt.xlabel("x-independiente")
  plt.ylabel("y-dependiente")
  plt.show()
```

```python
def main():
  #DATASET
  x=np.array([1,2,3,4,5])
  y=np.array([2,3,5,6,5])

  #Se obtiene b0 y b1
  b=estimate_b0_b1(x,y)
  print("Valores: b={}, b1={}".format(b[0],b[1]))
  
  plot_regression(x,y,b)
```

```python
if __name__=="__main__":
  main()
```
